```python
print("Assimetria: %f" % df_train['SalePrice'].skew()) #torção da curva com relação à distribuição normal (cauda londa à direita)
print("Curtose: %f" % df_train['SalePrice'].kurt()) #achatamento da curva com relação à distribuição normal (menos achatada - leptocúrtica)
```

Agora, avaliaremos as relações entre parâmetros numéricos:
    1. GrLivArea
    2. TotalBsmtSF

Para isso, usaremos o scatterplot.

```python
var = 'GrLivArea'

x = np.array(df_train[var])
y = np.array(df_train['SalePrice'])

plt.figure(figsize=(20,7))
plt.plot(x, y, 'o', c = 'red')
plt.xlabel(var)
plt.ylabel('SalePrice')
```

```python
var = 'TotalBsmtSF'

x = np.array(df_train[var])
y = np.array(df_train['SalePrice'])

plt.figure(figsize=(20,7))
plt.plot(x, y, 'o', c = 'green')
plt.xlabel(var)
plt.ylabel('SalePrice')
```

```python
df_train.head()
```

```python
temp = df_train['TotalBsmtSF'] == 0.0
aux_ids = []
for i in range(len(temp)):
    if temp[i] == True:
        aux_ids.append(i)
df_train = df_train.drop(aux_ids, axis = 0)
```

```python
var = 'TotalBsmtSF'

x = np.array(df_train[var])
y = np.array(df_train['SalePrice'])

plt.figure(figsize=(20,7))
plt.plot(x, y, 'o', c = 'green')
plt.xlabel(var)
plt.ylabel('SalePrice')
```

Agora, avaliaremos as relações entre parâmetros categóricos:
    1. OverallQual
    2. YearBuilt
    3. LotShape

Para isso, usaremos o boxplot.

```python
var = 'OverallQual'

data = pd.concat([df_train['SalePrice'], df_train[var]], axis=1)

f, ax = plt.subplots(figsize=(20, 7))
ax.set_ylim(0,800000)

sns.boxplot(x=var, y="SalePrice", data=data, notch = True)
```

```python
var = 'YearBuilt'
data = pd.concat([df_train['SalePrice'], df_train[var]], axis=1)
f, ax = plt.subplots(figsize=(20, 7))
fig = sns.boxplot(x=var, y="SalePrice", data=data)
fig.axis(ymin=0, ymax=800000);
```

```python
var = 'LotShape'
data = pd.concat([df_train['SalePrice'], df_train[var]], axis=1)
f, ax = plt.subplots(figsize=(20, 7))
fig = sns.boxplot(x=var, y="SalePrice", data=data)
fig.axis(ymin=0, ymax=800000);
```

Correlation Matrix: heatmap style

Com isso podemos avaliar casos de multilinearidade

```python
corrmat = df_train.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True);
```

Agora, removo atributos multilineares:

```python
df_train = df_train.drop(['1stFlrSF', '2ndFlrSF', 'GarageYrBlt', 'TotRmsAbvGrd', 'GarageCars', 'MasVnrArea'], axis = 1)
```

```python
corrmat = df_train.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True)
```

Agora, ampliaremos a matriz de correlação em torno do 'SalePrice':

```python
n_variables = 10
cols = corrmat.nlargest(n_variables, 'SalePrice')['SalePrice'].index
cm = np.corrcoef(df_train[cols].values.T)
sns.set(font_scale=1.25)

plt.figure(figsize=(20,7))
hm = sns.heatmap(cm, cbar = True, annot = True, square = True, fmt = '.2f', annot_kws = {'size' : n_variables}, yticklabels= cols.values, xticklabels=cols.values)
plt.show()
```

Agora farei multiplos scatter plots entre os 10 atributos acima que inclui 'SalePrice':

Agora trabalharemos com os dados faltantes (missing data)

```python
total = df_train.isnull().sum().sort_values(ascending = False)
percent = (df_train.isnull().sum()/df_train.isnull().count()).sort_values(ascending = False)
missing_data = pd.concat([total, percent], axis = 1, keys = ['Total', 'Percent'])
```

```python
missing_data.head(10)
```

Para atributos com mais de 15% de dados faltantes, os tomaremos como excludentes.

```python
missing_features = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu' , 'LotFrontage']
df_train = df_train.drop(missing_features, axis = 1)
```

```python
total = df_train.isnull().sum().sort_values(ascending = False)
percent = (df_train.isnull().sum()/df_train.isnull().count()).sort_values(ascending = False)
missing_data = pd.concat([total, percent], axis = 1, keys = ['Total', 'Percent'])
missing_data.head(10)
```

Outra forma de deletar:

```python
df_train = df_train.drop((missing_data[missing_data['Total'] > 1]).index, 1)
                                #deletando todos aqueles que nao possuem algum dado
df_train = df_train.drop(df_train.loc[df_train['Electrical'].isnull()].index, 0)
```

```python
total = df_train.isnull().sum().sort_values(ascending = False)
percent = (df_train.isnull().sum()/df_train.isnull().count()).sort_values(ascending = False)
missing_data = pd.concat([total, percent], axis = 1, keys = ['Total', 'Percent'])
missing_data.head(10)
```
